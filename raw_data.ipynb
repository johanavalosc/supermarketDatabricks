```python
df_spark = (spark.read
    .format("csv")
    .option("header", "true")        # Usa la primera fila como nombres de columnas
    .option("inferSchema", "true")   # Detecta automáticamente si es número o texto
    .load(f"{download_place}/supermarket.csv"))
limit = 20
df_spark = df_spark.limit(limit)
display(df_spark)
```

```python
for col_name in df_spark.columns:
    df_spark = df_spark.withColumnRenamed(col_name, col_name.replace(" ", "_"))
display(df_spark)
df_spark.write.mode("overwrite").saveAsTable("supermarket")
```
